```python
from autoscraper import AutoScraper
```

```python
Url='https://www.jumia.co.ke/mobile-phones/?tag=JMALL#catalog-listing'
wanted_list = [
    "Infinix Smart 8, 6.6\", 128GB + 4GB RAM, 5000mAh, 4G, DUAL SIM - Timber Black",
    "KSh 14,359",
    "KSh 17,999",
    "(21)"
]
```

```python
scraper=AutoScraper()
result=scraper.build(Url,wanted_list)
print(result)
```

```python
scraper.get_result_similar("https://www.jumia.co.ke/mobile-phones/?tag=JMALL#catalog-listing",grouped=True)
```

```python
scraper.set_rule_aliases({'rule_uew3':'Product Name','rule_msct':'Product Price','rule_9t70':'Price before Discount'})
scraper.keep_rules(['rule_uew3','rule_msct','rule_9t70'])
scraper.save("Jumia Search")
```

```python
result=scraper.get_result_similar("https://www.jumia.co.ke/mobile-phones/?tag=JMALL#catalog-listing",group_by_alias=True)
```
